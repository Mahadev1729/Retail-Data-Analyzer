```python
import pandas as pd
import numpy as np


df = pd.read_csv('sales_data.csv')


print(df.head())


print("\nMissing Values:\n", df.isnull().sum())


df[df.select_dtypes(include=['number']).columns] = df.select_dtypes(include=['number']).fillna(df.mean(numeric_only=True))


df['Total Sales'] = df['Units Sold'] * df['Price per Unit']
print("\nData with Total Sales:\n", df.head())


total_sales = np.sum(df['Total Sales'])
print(f"\nTotal Sales: ${total_sales}")


average_sales_per_product = df.groupby('Product')['Total Sales'].mean()
print("\nAverage Sales per Product:\n", average_sales_per_product)


best_selling_product = df.groupby('Product')['Total Sales'].sum().idxmax()
print(f"\nBest-Selling Product: {best_selling_product}")


df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.to_period('M')


monthly_sales = df.groupby('Month')['Total Sales'].sum()
print("\nMonthly Sales:\n", monthly_sales)


mean_sales = np.mean(df['Total Sales'])
median_sales = np.median(df['Total Sales'])
std_sales = np.std(df['Total Sales'])

print(f"\nMean Sales: ${mean_sales:.2f}")
print(f"Median Sales: ${median_sales:.2f}")
print(f"Standard Deviation of Sales: ${std_sales:.2f}")

df.to_csv('sales_data_analysis.csv', index=False)

print("\nSales data analysis has been saved to 'sales_data_analysis.csv'.")
```
